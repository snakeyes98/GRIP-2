```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv("SampleSuperStore.csv")
data.head(5)
```

```python
data.info()
```

```python

data.shape
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data['Sales'],color='green',label='Sales',shade=True,bw=25)
sns.kdeplot(data['Profit'],color='red',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

```python
sns.pairplot(data,hue='Category')
```

```python
sns.pairplot(data,hue='Region')
```

```python


plt.figure(figsize=(20,8))
sns.countplot(data['Sub-Category'])
plt.title('Sub-Category',fontsize=20)

```

```python
df2 = data['City'].value_counts()
df2=df2.head(50)
```

```python
df2.plot(kind='bar',figsize=(15,5))
plt.ylabel('Frequency / Number of deals')
plt.xlabel('City')

plt.title('City Wise Dealings', fontsize = 20)
plt.show()
```

```python
data['City'].value_counts().mean()
```

```python


df_state= data.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)

```

```python
df_state1=df_state.sort_values('Profit')
df_state1[['Profit']].plot(kind = 'bar', figsize = (10,10))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit ')
plt.xlabel('States')
plt.show()
# Vermont: Highest Profit
#  Ohio: Lowest Profit
```

```python
df_state1['Discount'].plot(kind='hist',figsize=(18,5))
plt.title('State wise analysis of Discount', fontsize=20)
```

```python
df_city= data.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
df_city = df_city.sort_values('Profit')
df_city.head(10)
```

```python
df_category = data.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
df_category
```

```python
data.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',
                                                              figsize= (10,10),
                                                              label=['Avg Sales Price($)','Profit($)'])
```
